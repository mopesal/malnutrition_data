```python
import pandas as pd
import numpy as np
mal_df = pd.ExcelFile('Desktop/Data science/Hamoye internship/malnutrition dataset.xlsx')
malnutrition = mal_df.parse(0, skiprows=3)
malnutrition.columns = ['country_code','country','year_period','median_year','start_month','end_month','WHO_ref_num','rep_level','region','other_grp','age',\
                        'sex','mother_education','wealth_quintile','urban_rural','sample_size','severe_wasting','wasting','overweight','stunting','underweight',\
                        'short_source_code','reference_title','author','notes','JME(Y/N)']
print(malnutrition.info())
print(malnutrition.isna().sum())
#print(malnutrition.country.count())
```

```python
malnutrition['start_month'].str.strip()
malnutrition['start_month'] = malnutrition['start_month'].str.title()
malnutrition['start_month'] = malnutrition['start_month'].str.replace('January','Jan')
malnutrition['start_month'] = malnutrition['start_month'].str.replace('February','Feb')
malnutrition['start_month'] = malnutrition['start_month'].str.replace('March','Mar')
malnutrition['start_month'] = malnutrition['start_month'].str.replace('April','Apr')
malnutrition['start_month'] = malnutrition['start_month'].str.replace('June','Jun')
malnutrition['start_month'] = malnutrition['start_month'].str.replace('July','Jul')
malnutrition['start_month'] = malnutrition['start_month'].str.replace('August','Aug')
malnutrition['start_month'] = malnutrition['start_month'].str.replace('September','Sep')
malnutrition['start_month'] = malnutrition['start_month'].str.replace('October','Oct')
malnutrition['start_month'] = malnutrition['start_month'].str.replace('November','Nov')
malnutrition['start_month'] = malnutrition['start_month'].str.replace('December','Dec')
malnutrition['start_month'] = malnutrition['start_month'].str.replace('Juln','Jul')
malnutrition['start_month'] = malnutrition['start_month'].str.replace('Nmar','Mar')
malnutrition['start_month'] = malnutrition['start_month'].fillna('Sep') 

#
malnutrition['end_month'] = malnutrition['end_month'].str.title()
malnutrition['end_month'] = malnutrition['end_month'].fillna('Dec')
#
malnutrition['rep_level'] = malnutrition['rep_level'].fillna('R')
#
malnutrition['mother_education'] = malnutrition['mother_education'].fillna('unknown')
#
malnutrition['wealth_quintile'] = malnutrition['wealth_quintile'].fillna('unclassified')
#
malnutrition['region'] = malnutrition['region'].fillna('unknown')
#
malnutrition['sample_size'] = malnutrition['sample_size'].fillna(malnutrition['sample_size'].mean())
#
malnutrition['wasting'] = malnutrition['wasting'].fillna(malnutrition['wasting'].mean())
#
malnutrition['severe_wasting'] = malnutrition['severe_wasting'].fillna(malnutrition['severe_wasting'].mean())
#
malnutrition['overweight'] = malnutrition['overweight'].fillna(malnutrition['overweight'].mean())
#
malnutrition['stunting'] = malnutrition['stunting'].fillna(malnutrition['stunting'].mean())
#
malnutrition['underweight'] = malnutrition['underweight'].fillna(malnutrition['underweight'].mean())
#
malnutrition['notes'] = malnutrition['notes'].fillna('unavailable')
#
malnutrition['other_grp'] = malnutrition['other_grp'].fillna('uncategorized')
#

print(malnutrition.isna().sum())
```
